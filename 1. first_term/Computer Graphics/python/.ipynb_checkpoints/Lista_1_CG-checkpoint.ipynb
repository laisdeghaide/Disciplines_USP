#Exercício 2


## *A matriz de transformação da translação é dada da seguinte forma:*

><center>$T=\begin{bmatrix}1 & 0 & tx\\
0 & 1 &  ty\\
0 & 0 & 1\\
\end{bmatrix}$

<br><br>

## *Já a matriz de transformação da rotação é dada da seguinte forma:*

><center>$R=\begin{bmatrix}cos(\theta) & -sen(\theta) & 0\\
sen(\theta) & cos(\theta) &  0\\
0 & 0 & 1\\
\end{bmatrix}$

<br><br>

## *Portanto, ao multiplicar as duas matrizes, temos:*

><center>$T*R=\begin{bmatrix}cos(\theta) & -sen(\theta) & tx\\
sen(\theta) & cos(\theta) &  ty\\
0 & 0 & 1\\
\end{bmatrix}$

#Exercício 3

## *A matriz de transformação da translação é dada da seguinte forma:*

><center>$T=\begin{bmatrix}1 & 0 & tx\\
0 & 1 &  ty\\
0 & 0 & 1\\
\end{bmatrix}$

<br><br>

Substituindo:

><center>$T=\begin{bmatrix}1 & 0 & 11\\
0 & 1 &  27\\
0 & 0 & 1\\
\end{bmatrix}$

<br><br>


## *Já a matriz de transformação da escala é dada da seguinte forma:*

><center>$S=\begin{bmatrix}sx & 0 & 0\\
0 & sy &  0\\
0 & 0 & 1\\
\end{bmatrix}$

<br><br>

Substituindo:

><center>$S=\begin{bmatrix}2 & 0 & 0\\
0 & 2 &  0\\
0 & 0 & 1\\
\end{bmatrix}$

<br><br>


## *Portanto, ao multiplicar as duas matrizes, temos:*

><center>$T*S=\begin{bmatrix}2 & 0 & 11\\
0 & 2 &  27\\
0 & 0 & 1\\
\end{bmatrix}$

#Exercício 4

## *Já a matriz de transformação de rotação é dada da seguinte forma:*


><center>$R(27+11)=\begin{bmatrix}cos(27+11) & -sen(27+11) & 0\\
sen(27+11) & cos(27+11) &  0\\
0 & 0 & 1\\
\end{bmatrix}$


><center>$R(38)=\begin{bmatrix}cos(38) & -sen(38) & 0\\
sen(38) & cos(38) &  0\\
0 & 0 & 1\\
\end{bmatrix}$


><center>$R(38)=\begin{bmatrix}0.79 & -0.62 & 0\\
0.62 & 0.79 &  0\\
0 & 0 & 1\\
\end{bmatrix}$

<br><br>


><center>$R(27)=\begin{bmatrix}cos(27) & -sen(27) & 0\\
sen(27) & cos(27) &  0\\
0 & 0 & 1\\
\end{bmatrix}$



><center>$R(11)=\begin{bmatrix}cos(11) & -sen(11) & 0\\
sen(11) & cos(11) &  0\\
0 & 0 & 1\\
\end{bmatrix}$



><center>$R(27)*R(11)=\begin{bmatrix}cos(27)cos(11)-sen(27)sen(11) & -cos(27)sen(11)-sen(27)cos(11) & 0\\
sen(27)cos(11)+cos(27)sen(11) & -sen(27)sen(11)+cos(27)cos(11) &  0\\
0 & 0 & 1\\
\end{bmatrix}$

><center>$R(27)*R(11)=\begin{bmatrix}0.79 & -0.62 & 0\\
0.62 & 0.79 &  0\\
0 & 0 & 1\\
\end{bmatrix}$
